```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
###loading model
model=tf.keras.models.load_model('trained_model1.keras')
```

```python
###visualizing single image of test set

import cv2

image_path="Testing/glioma_tumor/image(38).jpg"
#reading image

img=cv2.imread(image_path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  #convert bgr to rgb


#displaying image
plt.imshow(img)
plt.title("test image")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
##testing model

image=tf.keras.preprocessing.image.load_img(image_path,target_size=(225,225))
input_arr=tf.keras.preprocessing.image.img_to_array(image)
input_arr=np.array([input_arr])  #convert single image to a batch
print(input_arr.shape)
```

```python
prediction=model.predict(input_arr)
prediction,prediction.shape
```

```python
result_index=np.argmax(prediction)
result_index
```

```python
class_name=['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']
```

```python
model_prediction=class_name[result_index]
plt.imshow(img)
plt.imshow(img)
plt.title(f"disease name:{model_prediction}")
plt.xticks([])
plt.yticks([])
plt.show()
```
